```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels
from dateutil.parser import parse
import os 

landing = r'D:\_workspace\tcc\CommodityPriceForecast\data\LandingZone\IMF'
raw = r'D:\_workspace\tcc\CommodityPriceForecast\data\RawZone'
file = 'Primary_Commodity_Price_System_PCPS-1.xlsx'

data_out = 'data'
md_out = 'metadata'

r_path = os.path.join(landing,file)
```

```python
data = pd.read_excel(r_path, sheet_name=0) # Data Sheet
md = pd.read_excel(r_path,sheet_name=1) # Meta Data Sheet
```

```python
#################
#Transpose Data
################

# transpose
df = data.T

headers = df.iloc[0]
df.columns = [str.lower(str(c)) for c in headers] 
df = df[df.columns[:-2]].iloc[2:]
df.reset_index(drop=False, inplace=True)
df = df.rename(columns={
    'index': 'month'
})
#df.set_index('month', inplace=True)
df.head()
```

```python
############################
#Extract Datetime Features
###########################

df['dt_ym'] = pd.to_datetime(df['month'])

df['dt_y'] = pd.DatetimeIndex(df['dt_ym']).year
df['dt_m'] = pd.DatetimeIndex(df['dt_ym']).month
df.head()
```

```python
######################
#Transpose Meta Data
######################

df_md = md.T

headers = df_md.iloc[0]
df_md.columns = [str.lower(str(c)) for c in headers] 
df_md = df_md[df_md.columns[:-2]].iloc[2:]
df_md.reset_index(drop=False, inplace=True)
df_md = df_md.rename(columns={
    'index': 'month'
})

#df_md['dt_ym'] = pd.to_datetime(df_md['month'])
#df_md.set_index(df_md.month, drop=True, inplace=True)
#df_md['dt_y'] = pd.DatetimeIndex(df_md['dt_ym']).year
#df_md['dt_m'] = pd.DatetimeIndex(df_md['dt_ym']).month

#df_md.drop(columns=['month'], inplace=True)

df_md.set_index('month', inplace=True)
df_md
```

```python
df_md = df_md.replace(to_replace='\n.*', value='',regex=True)

md_dict = {}

for c in df_md.columns:
    md_dict[c] = df_md[c].unique()

pd.DataFrame(md_dict)
```

```python
# save to raw 

df.to_csv(os.path.join(raw,'IMF','data.csv'), index=False)

pd.DataFrame(md_dict).to_csv(os.path.join(raw,'IMF','masterdata.csv'))
```
